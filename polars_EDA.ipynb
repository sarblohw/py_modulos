```python
import numpy as np
import pandas as pd
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
cf.go_offline()
%matplotlib inline
pd.options.display.max_columns = 100
pd.options.display.max_rows = 25
pd.options.display.precision = 2
```

```python
dfpl = pl.DataFrame(sns.load_dataset('mpg'))
```

```python
dfpl.head()
```

```python
dfpl\
.select(pl.all())\
.filter((pl.col('model_year') >= 70) | (pl.col('model_year') <= 75))\
.shape
```

```python
dfpl\
.filter((pl.col('model_year') >= 70) | (pl.col('model_year') <= 75))\
.select(['displacement', 'horsepower', 'weight', 'acceleration', 'mpg'])
```

```python
dfpl\
.filter((pl.col('model_year') >= 70) | (pl.col('model_year') <= 75))\
.groupby(['origin', 'model_year', 'cylinders'])\
.agg(pl.all().mean())
```

```python
dfpl\
.filter((pl.col('model_year') >= 70) | (pl.col('model_year') <= 75))\
.to_pandas()\
.drop(columns = 'name')\
.groupby(['origin', 'model_year', 'cylinders']).agg(['mean', 'count', 'sum'])
```

```python
dfpl\
.filter((pl.col('model_year') >= 70) | (pl.col('model_year') <= 75))\
.to_pandas()\
.drop(columns = 'name')\
.groupby(['origin', 'model_year', 'cylinders'])\
.get_group(('usa', 70, 6))\
.agg(['mean', 'count'])
```
